```python
import os
import csv
```

```python
dataDir = '/Users/cameronhaynes/Documents/VSCodeGitProjects/AvidaGeneDupe/experiments/2022-2-4-PaperDuplicationWMod/testResults'
runDirectories = []
for subdir in os.listdir(dataDir):
    treatmentDir = os.path.join(dataDir,subdir)
    for run_dir in os.listdir(treatmentDir):
        if not 'run_' in run_dir:
            continue
        runDirectories.append(os.path.join(treatmentDir,run_dir))
runDirectories
```

```python
#Use r"Path" to avoid any problems from special characters
def getOrganisms(filePath):
    with open(filePath,'r') as datFile:
        lines = datFile.readlines()
        initalOrgPos = 0
        for k,line in enumerate(lines):
            if (line[0] != '') & (line[0] != '#') & (line[0] != '\n'):
                initialOrgPos = k
                break
            else:
                continue

        organisms = []
        for i in range(initialOrgPos,len(lines)):
            if(lines[i] != ''):
                organisms.append(lines[i])
            else:
                continue

        if(len(organisms) > 0):
            return organisms
        else:
            print("Error: please check code")
```

```python
#Does os.path.join automatically insert necessary /? Yes!
dataFile = os.path.join(runDirectories[0],"data/detail_MostNumLineage.dat")
organisms = getOrganisms(dataFile)
for org in organisms:
    print(org)
```

```python
#List slicing goes up to but does not include the last element!!
def countTasks(organismString, isLatterPartOfLineage):
    presentTasks = 0
    
    analyzeOutputs = organismString.split()
    if(isLatterPartOfLineage==False):
        tasks = analyzeOutputs[2:11]
    else:
        tasks = analyzeOutputs[3:12]
        
    for task in tasks:
        if int(task) != 0:
            presentTasks +=1
    return presentTasks
```

```python
countTasks(organisms[1],True)
```

```python
def getOrganismID(organismString):
    analyzeOutputs = organismString.split()
    ID = analyzeOutputs[0]
    return ID

def getUpdateBorn(organismString):
    analyzeOutputs = organismString.split()
    updateBorn = analyzeOutputs[1]
    return updateBorn
```

```python
print(getOrganismID(organisms[0]))
print(getUpdateBorn(organisms[0]))
```

```python
def makeTidy(lineageData,fullLineage):
    datFile = os.path.join(lineageData[0],lineageData[1])
    organisms = getOrganisms(datFile)
    
    with open('{}.csv'.format(lineageData[0]), mode='w') as tidyDat:
        data_writer = csv.writer(tidyDat, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        data_writer.writerow(['Organism ID','Observation','Value'])
        
        for organism in organisms:
            ID = getOrganismID(organism)
            updateBorn = getUpdateBorn(organism)
            tasks = countTasks(organism,fullLineage)
            data_writer.writerow([ID,"Update Born",updateBorn])
            data_writer.writerow([ID,"Task Count",tasks])
```

```python
dataFolder = os.path.join(runDirectories[0],"data")
linDatFile = "detail_MostNumLineage.dat"
#Data Name, format
linData = [dataFolder,linDatFile]
makeTidy(linData,True)
```
